```python
import pandas as pd
```

```python
crime = pd.read_csv('경찰청_범죄 발생 지역별 통계_20231231.csv', encoding='euc-kr')
```

```python
busan_crime = crime[['범죄중분류','부산중구','부산서구','부산동구','부산영도구','부산부산진구','부산동래구','부산남구','부산북구','부산강서구','부산해운대구','부산사하구','부산금정구','부산연제구','부산수영구','부산사상구','부산기장군']]
```

```python
busan_crime.head()
```

```python
busan_crime.set_index('범죄중분류', inplace=True)

busan_crime = busan_crime.transpose()
```

```python
busan_crime.head()
```

```python
busan_crime.columns
```

```python
busan_crime = busan_crime.iloc[:,:17]
busan_crime.drop(['공갈','손괴','방화','절도범죄'], axis=1, inplace=True)
```

```python
busan_crime['범죄합계']= busan_crime.sum(axis=1)
```

```python
busan_crime
```

```python
busan_crime.to_csv('busan_crime_analysis.csv', encoding='utf-8-sig')
```

```python
import numpy as np
import matplotlib.pyplot as plt

for c in ['종합치안점수']:

# 데이터 정렬
    busan_crime_concat.sort_values(c, ascending=False, inplace=True)

# 색상 지정
    colors1 = ["skyblue"] * len(busan_crime_concat)
    colors2 = ["lightcoral"] * len(busan_crime_concat)


    x = np.arange(len(busan_crime_concat))  # 도시 개수만큼 x좌표
    width = 0.4               # 막대 너비

    fig, ax1 = plt.subplots(figsize=(10,6))

# 보안등 막대
    bars1 = ax1.bar(x - width/2, busan_crime_concat[c], width, color=colors1, label=c)
# 범죄율 막대
    ax2=ax1.twinx()
    bars2 = ax2.bar(x + width/2, busan_crime_concat['합계'], width, color=colors2, label="합계 범죄율")
    ax2.set_ylabel("범죄율 (만명당 건수)")
# 보안등 값 표시
    for bar in bars1:
        h = bar.get_height()
        ax1.text(bar.get_x() + bar.get_width()/2, h, f"{round(h,1):,}", 
             ha='center', va='bottom', fontsize=8)

# 범죄율 값 표시
    for bar in bars2:
        h = bar.get_height()
        ax2.text(bar.get_x() + bar.get_width()/2, h, f"{round(h,1):,}", 
             ha='center', va='bottom', fontsize=8)

# x축 설정
    ax1.set_xticks(x)
    ax1.set_xticklabels(busan_crime_concat.index, rotation=45)
    ax1.set_title(f"도시별 {c} vs 범죄율 비교")
    ax1.set_xlabel("도시")
    ax1.set_ylabel("수치")
    ax1.legend()

    plt.tight_layout()
    
    plt.show()
    
```
