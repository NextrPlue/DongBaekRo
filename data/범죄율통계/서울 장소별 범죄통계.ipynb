```python
import pandas as pd
df = pd.read_csv('경찰청 서울특별시경찰청_5대범죄 발생 장소별 현황_20231231.csv',encoding = 'euc-kr')  # Example placeholder for context
```

```python
df.head()
```

```python
df = df[df["발생건수"] > 1]
```

```python
murder = df[df["범죄명"] == "살인"]
```

```python
import matplotlib.pyplot as plt
import koreanize_matplotlib
```

```python
plt.figure(figsize=(10,6))
bars = plt.bar(murder["장소"], murder["발생건수"], color="skyblue")

# 값 표시
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, height, str(height),
             ha="center", va="bottom")

plt.title("장소별  살인 발생 건수")
plt.xlabel("장소")
plt.ylabel("건수")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()
```

```python
chu= df[df["범죄명"] == "강간·강제추행"]
```

```python
chu[chu['장소'] == '단독주택'].index[0]
```

```python
import numpy as np
import matplotlib.pyplot as plt
targets_orange = ['단독주택', '아파트', '다세대/연립', '오피스텔(원룸)', '기타거주시설(기숙사 등)']
colors = ["skyblue"] * len(chu)
chu.reset_index(drop=True, inplace=True)
for i in chu[chu['장소'].isin(targets_orange)].index:
    colors[i] = 'orange'
for i in chu[chu['장소'].eq('통행로(보도/골목길)')].index:
    colors[i] = 'blue'
plt.figure(figsize=(10,6))


bars = plt.bar(chu["장소"], chu["발생건수"], color=colors)

# 값 표시
for bar in bars:
    h = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, h, f"{h}", ha="center", va="bottom")

plt.title("장소별  강간 강제추행 발생 건수")
plt.xlabel("장소")
plt.ylabel("건수")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()
```
